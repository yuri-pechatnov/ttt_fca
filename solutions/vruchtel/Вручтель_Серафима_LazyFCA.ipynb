#### Нужен one_hot encoding

```python
def one_hot_encode(df):

    def apply_num(letter, x):
        return 1 if x == letter else 0
        
    df = (df.assign(V1_x = lambda df: df['V1'].apply(lambda x: apply_num('x', x)),
                    V1_o = lambda df: df['V1'].apply(lambda x: apply_num('o', x)),
                    V1_b = lambda df: df['V1'].apply(lambda x: apply_num('b', x)),
                    
                    V2_x = lambda df: df['V2'].apply(lambda x: apply_num('x', x)),
                    V2_o = lambda df: df['V2'].apply(lambda x: apply_num('o', x)),
                    V2_b = lambda df: df['V2'].apply(lambda x: apply_num('b', x)),
                    
                    V3_x = lambda df: df['V3'].apply(lambda x: apply_num('x', x)),
                    V3_o = lambda df: df['V3'].apply(lambda x: apply_num('o', x)),
                    V3_b = lambda df: df['V3'].apply(lambda x: apply_num('b', x)),
                    
                    V4_x = lambda df: df['V4'].apply(lambda x: apply_num('x', x)),
                    V4_o = lambda df: df['V4'].apply(lambda x: apply_num('o', x)),
                    V4_b = lambda df: df['V4'].apply(lambda x: apply_num('b', x)),
                    
                    V5_x = lambda df: df['V5'].apply(lambda x: apply_num('x', x)),
                    V5_o = lambda df: df['V5'].apply(lambda x: apply_num('o', x)),
                    V5_b = lambda df: df['V5'].apply(lambda x: apply_num('b', x)),
                    
                    V6_x = lambda df: df['V6'].apply(lambda x: apply_num('x', x)),
                    V6_o = lambda df: df['V6'].apply(lambda x: apply_num('o', x)),
                    V6_b = lambda df: df['V6'].apply(lambda x: apply_num('b', x)),
                    
                    V7_x = lambda df: df['V7'].apply(lambda x: apply_num('x', x)),
                    V7_o = lambda df: df['V7'].apply(lambda x: apply_num('o', x)),
                    V7_b = lambda df: df['V7'].apply(lambda x: apply_num('b', x)),
                    
                    V8_x = lambda df: df['V8'].apply(lambda x: apply_num('x', x)),
                    V8_o = lambda df: df['V8'].apply(lambda x: apply_num('o', x)),
                    V8_b = lambda df: df['V8'].apply(lambda x: apply_num('b', x)),
                    
                    V9_x = lambda df: df['V9'].apply(lambda x: apply_num('x', x)),
                    V9_o = lambda df: df['V9'].apply(lambda x: apply_num('o', x)),
                    V9_b = lambda df: df['V9'].apply(lambda x: apply_num('b', x)),
                    
                    V10_bin = lambda df: df['V10'].apply(lambda x: 1 if x == 'positive' else 0)
                   )
         )
        
    return df.drop(columns=['V' + str(col_id) for col_id in range(1, 11)])
```

```python
attrib_names = [ '1_x', '1_o', '1_b', '2_x', '2_o', '2_b', '3_x', '3_o', '3_b',
                 '4_x', '4_o', '4_b', '5_x', '5_o', '5_b', '6_x', '6_o', '6_b',
                 '7_x', '7_o', '7_b', '8_x', '8_o', '8_b', '9_x', '9_o', '9_b']

def make_intent(element):
    return set([i+':'+str(k) for i,k in zip(attrib_names, element[:-1])])
```

```python
%%time
def calculate_support(context_plus, element):
    """
    Вычисляет поддержку для элемента element
    Поддержка: количество положительных примеров, которые содержат (element intersect plus_i) / количество положительных примеров
    """
    one_el_supports = []
    intent = make_intent(element)
    
    context_plus_intents = [make_intent(plus_el) for index, plus_el in context_plus.iterrows()]
    intersects_intents = [intent & plus_intent for plus_intent in context_plus_intents]
    
    plus_elements_count = context_plus.shape[0]
    
    for intersect_intent in intersects_intents:
        current_count = 0
        
        for plus_intent in context_plus_intents:
            if plus_intent.issuperset(intersect_intent):
                current_count += 1
                
        one_el_supports.append(current_count / plus_elements_count)
        
    return one_el_supports

def calculate_confidence(context_plus, context_minus, element):
    """
    Вычисляет что-то вроде достоверности для элемента element:
    количество отрицательных примеров, которые содержат (element intersect plus_i) / кол-во отрицательных примеров
    """
    one_el_confidences = []
    intent = make_intent(element)
    
    intersects_intents = [intent & make_intent(plus_el) for index, plus_el in context_plus.iterrows()]
    context_minus_intents = [make_intent(minus_el) for index, minus_el in context_minus.iterrows()]
    
    minus_elements_count = context_minus.shape[0]
    
    for intersect_intent in intersects_intents:
        current_count = 0
        
        for minus_intent in context_minus_intents:
            if minus_intent.issuperset(intersect_intent):
                current_count += 1
        one_el_confidences.append(current_count / minus_elements_count)
    return one_el_confidences

support = calculate_support(context_plus_tmp, bin_test_df.iloc[0])
print(len(support))
print(support)
print()

confidence = calculate_confidence(context_plus_tmp, context_minus_tmp, bin_test_df.iloc[0])
print(len(confidence))
print(confidence)
print()
```

```python
def predict_by_support_and_confidence(support_list, confidence_list, support_bound, confidence_bound, if_equal):
    """
    if_equal - возвращаемое значение в случае одинакового количества голосов (voices_sum = 0)
    """
    assert(len(support_list) == len(confidence_list))
    voices_sum = 0
    # +1, если support выше support_bound, -1, если confidence ниже confidence_bound
    for i in range(len(support_list)):
        if support_list[i] >= support_bound:
            voices_sum += 1
        if confidence_list[i] >= confidence_bound:
            voices_sum -= 1
    if voices_sum > 0:
        return 1
    if voices_sum < 0:
        return 0
    return if_equal

predict_by_support_and_confidence(support, confidence, support_bound=0.01, confidence_bound=0.01, if_equal=1)
```

```python
class Model:
    
    def __init__(self, bin_train_df, bin_test_df, support_bound, confidence_bound, if_equal):
        self.context_plus = bin_train_df.loc[bin_train_df.V10_bin == 1]
        self.context_minus = bin_train_df.loc[bin_train_df.V10_bin == 0]
        self.test_elements = bin_test_df
        self.test_targets = bin_test_df['V10_bin'].values        
        
        self.support_bound = support_bound
        self.confidence_bound = confidence_bound
        self.if_equal = if_equal
    
    def make_list_of_intents(self, elements):
        return [make_intent(el) for el in elements]
    
    def lazy_fit(self):
        self.context_plus_intents = self.make_list_of_intents(self.context_plus.values)
        self.context_minus_intents = self.make_list_of_intents(self.context_minus.values)
        self.test_intents = self.make_list_of_intents(self.test_elements.values)
    
    def intersect_intent_with(self, intent, intents_list):
        return [intent & el_intent for el_intent in intents_list]
    
    def calculate_support(self, intent, intersects_intents):
        """
        Вычисляет поддержку для элемента element
        Поддержка: количество положительных примеров, которые содержат (element intersect plus_i) 
            / количество положительных примеров
        """
        one_el_supports = []
        plus_elements_count = self.context_plus.shape[0]
    
        for intersect_intent in intersects_intents:
            current_count = 0

            for plus_intent in self.context_plus_intents:
                if plus_intent.issuperset(intersect_intent):
                    current_count += 1
            one_el_supports.append(current_count / plus_elements_count)

        return one_el_supports
    
    def calculate_confidence(self, intent, intersects_intents):
        """
        Вычисляет что-то вроде достоверности для элемента element:
        количество отрицательных примеров, которые содержат (element intersect plus_i) / кол-во отрицательных примеров
        """
        one_el_confidences = []
        minus_elements_count = self.context_minus.shape[0]

        for intersect_intent in intersects_intents:
            current_count = 0

            for minus_intent in self.context_minus_intents:
                if minus_intent.issuperset(intersect_intent):
                    current_count += 1
            one_el_confidences.append(current_count / minus_elements_count)
            
        return one_el_confidences

    
    def predict_one(self, element):
        """
        support_bound - значение, выше которого должно быть значение support, чтобы оно было учтено в голосовании
        аналогично confidence_bound
        if_equal - возвращаемое значение в случае одинакового количества голосов (voices_sum = 0)
        """
        intent = make_intent(element)
        
        intersects_intents = self.intersect_intent_with(intent, self.context_plus_intents)
        
        support_list = self.calculate_support(intent, intersects_intents)
        confidence_list = self.calculate_confidence(intent, intersects_intents)
        
        voices_sum = 0
        # +1, если support выше support_bound, -1, если confidence ниже confidence_bound
        for i in range(len(support_list)):
            if support_list[i] >= self.support_bound:
                voices_sum += 1
            if confidence_list[i] >= self.confidence_bound:
                voices_sum -= 1
        if voices_sum > 0:
            return 1
        if voices_sum < 0:
            return 0
        return self.if_equal

    def predict(self, elements):        
        return [self.predict_one(el) for el in elements]
        
    def metrics(self, real_targets, predict_targets):
        predictions_count = len(predict_targets)
        
        accuracy = accuracy_score(real_targets, predict_targets)
        precision = precision_score(real_targets, predict_targets) 
        recall = recall_score(real_targets, predict_targets)        
        
        TP = sum(np.array(real_targets) & np.array(predict_targets))
        FP = sum([1 if predict_targets[i] == 1 and real_targets[i] == 0 else 0 for i in range(predictions_count)])
        TN = sum([1 if predict_targets[i] == 0 and real_targets[i] == 0 else 0 for i in range(predictions_count)])
        FN = sum([1 if predict_targets[i] == 0 and real_targets[i] == 1 else 0 for i in range(predictions_count)])
        
        TPR = TP / (TP + FN) if (TP + FN) != 0 else np.inf  # true positive rate
        TNR = TN / (TN + FP) if (TN + FP) != 0 else np.inf  # true negarive rate
        
        NPV = TN / (TN + FN) if (TN + FN) != 0 else np.inf # negative predictive value
        
        FPR = FP / (FP + TN) if (FP + TN) != 0 else np.inf # false positive rate
        
        FDR = FP / (FP + TP) if (FP + TP) != 0 else np.inf  # false discovery rate
        
        return {'accuracy': accuracy, 'precision': precision, 'recall': recall,
               'TP': TP, 'FP': FP, 'TN': TN, 'FN': FN, 'TPR': TPR, 'FPR': FPR, 
               'true_negative_rate':TNR, 'negative_predictive_value':NPV, 'false_discovery_rate':FDR}
        
```

```python
def cross_validation(bin_train_df, support_bound, confidence_bound, if_equal, validation_part = 0.2):
    validation_start_idx = 0
    step = int(bin_train_df.shape[0] * validation_part)
    validation_end_idx = step
    
    metrics = []
    
    while validation_start_idx < bin_train_df.shape[0]:
        print(validation_start_idx, validation_end_idx)
        train_idxs = list(range(validation_start_idx)) + list(range(validation_end_idx,bin_train_df.shape[0]))
        test_idxs = list(range(validation_start_idx, validation_end_idx))
        
        cur_test_df = bin_train_df.iloc[test_idxs]
        model = Model(bin_train_df.iloc[train_idxs], cur_test_df, support_bound, confidence_bound, if_equal)
        model.lazy_fit()
        pred_valid = model.predict(cur_test_df.values)
        metrics.append(model.metrics(model.test_targets, pred_valid))
        
        validation_start_idx = validation_end_idx
        validation_end_idx += step
        if validation_end_idx > bin_train_df.shape[0]:
            validation_end_idx = bin_train_df.shape[0]
    
    return metrics

result_metrix = cross_validation(bin_train_df[:1000], support_bound=0.01, confidence_bound=0.01, if_equal=1, validation_part=0.2)
print(result_metrix)
```

```python
%%time
support_bounds = [0.005, 0.01, 0.02, 0.03]
confidence_bounds = [0.005, 0.01, 0.02, 0.03]
if_equal_var = [1, 0]

metrics = []

test_random_idx = random.sample(range(bin_test_df.shape[0]), 100)
train_random_idx = random.sample(range(bin_train_df.shape[0]), 1000)

for support_bound in support_bounds:
    for confidence_bound in confidence_bounds:
        for if_equal in if_equal_var:
            
            print("support_bound:", support_bound, "confidence_bound:", confidence_bound, "if_equal:", if_equal)
            model = Model(bin_train_df.iloc[train_random_idx], bin_test_df.iloc[test_random_idx], 
                          support_bound=support_bound, confidence_bound=confidence_bound, if_equal=if_equal)
            model.lazy_fit()
            predict_test = model.predict(bin_test_df.values[test_random_idx])
            cur_metrics = model.metrics(model.test_targets, predict_test)
            print(cur_metrics)
            metrics.append(cur_metrics)
```
